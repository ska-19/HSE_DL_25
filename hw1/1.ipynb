```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt

import numpy as np

import warnings
for warn in [UserWarning, FutureWarning]: warnings.filterwarnings("ignore", category = warn)
```

```python
image_path = "example.png"
original_image = Image.open(image_path)
```

```python
def plot_image(image, title="Augmented"):
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()

plot_image(original_image, "Original")
```

```python
augmentations = transforms.Compose([
    transforms.Resize((256, 256)), # Чтобы уменьшить колличество пикселей (сжать)
    transforms.ColorJitter(brightness=0.2, contrast=0.1, saturation=0.1, hue=0.1),
    transforms.RandomHorizontalFlip(p=0.5),  # 50% вероятность отражения
    transforms.RandomRotation(5),  # поворот на угол до 5 градусов
    transforms.GaussianBlur(kernel_size=5, sigma=(0.1, 2.0)),
])

augmented_image = augmentations(original_image)

plot_image(augmented_image)
```

```python
# 1. Resize
resize = transforms.Compose([
    transforms.Resize((256, 256)),
])
resized_image = resize(original_image)
plot_image(resized_image, "Resized Image")

# 2. ColorJitter
color_jitter = transforms.Compose([
    transforms.ColorJitter(brightness=0.2, contrast=0.1, saturation=0.1, hue=0.1),
])
color_jittered_image = color_jitter(original_image)
plot_image(color_jittered_image, "ColorJitter Image")

# 3. RandomHorizontalFlip
random_horizontal_flip = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
])
flipped_image = random_horizontal_flip(original_image)
plot_image(flipped_image, "RandomHorizontalFlip Image")

# 4. RandomRotation
random_rotation = transforms.Compose([
    transforms.RandomRotation(5),
])
rotated_image = random_rotation(original_image)
plot_image(rotated_image, "RandomRotation Image")

# 5. GaussianBlur
gaussian_blur = transforms.Compose([
    transforms.GaussianBlur(kernel_size=5, sigma=(0.1, 2.0)),
])
blurred_image = gaussian_blur(original_image)
plot_image(blurred_image, "GaussianBlur Image")
```

### Выводы

1. **Resize**: Изменение размера изображения позволяет стандартизировать размеры изображений, что важно для входных данных.
2. **ColorJitter**: Изменение яркости, контраста и насыщенности помогает модели лучше адаптироваться к цветовым вариациям.
3. **RandomHorizontalFlip**: Отражение изображения по горизонтали увеличивает количество данных.
4. **RandomRotation**: Поворот изображения на небольшой угол помогает лучше распознавать объекты под углом.
5. **GaussianBlur**: Размытие изображения может помочь модели стать более устойчивой к шумам и мелким деталям.

Эти аугментации помогают улучшить вариативность данных и повысить устойчивость модели к различным изменениям в изображениях и научиться проходить капчи :)
